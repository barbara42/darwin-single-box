```python
def get_single_value(file_name):
    with open(file_name, mode='rb') as file: # b is important -> binary
        file_content = file.read()
    bin_contents = struct.unpack(">" + "i" * (len(file_content) // 4), file_content)
    return bin_contents[0]
```
